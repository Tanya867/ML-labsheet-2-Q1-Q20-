```python


import pandas as pd

df = pd.DataFrame({"Name": ["Tanya", "Utkarsh", "Dhiraj"],
                   "Age": [25, 30, 35]})

filtered = df[df["Age"] > 28]
print("Filtered rows (Age > 28):\n", filtered)
```

```python


import pandas as pd

df = pd.DataFrame({"Name": ["Tanya", "Utkarsh", "Dhiraj"],
                   "Age": [25, 30, 35]})

df["Age+5"] = df["Age"] + 5
print("DataFrame with new column:\n", df)
```

```python

import pandas as pd

data = {"City": ["Delhi", "Delhi", "Mumbai", "Mumbai"],
        "Sales": [200, 250, 400, 300]}

df = pd.DataFrame(data)
grouped = df.groupby("City").mean()
print("Grouped average:\n", grouped)
```

```python


import pandas as pd
import matplotlib.pyplot as plt

categories = ["A", "B", "C", "A", "B", "C", "A"]
df = pd.DataFrame({"Category": categories})

df["Category"].value_counts().plot(kind="bar")
plt.title("Category Frequency")
plt.show()
```
